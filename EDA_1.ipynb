```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('data_clean.csv')
data1
```

```python
pd.set_option('display.min_rows',None)
data1
```

```python
type(data1)
data1.dtypes
data1.info()
```

```python
data2=data1.iloc[:,1:]
data2
```

```python
data=data2.copy()
pd.set_option('display.min_rows',None)
data
```

```python
data['Temp']=pd.to_numeric(data['Temp'],errors='coerce')
data.dtypes
#convert non numeric data to null
```

```python
data['Weather']=data['Weather'].astype('category')
data['Wind']=data['Wind'].astype('int64')
data.dtypes
#astype is used for changing dtype
```

```python
data[data.duplicated()].shape
```

```python
data[data.duplicated()]
```

```python
data_cleaned1=data.drop_duplicates()
data_cleaned1.shape
```

```python
data_cleaned3=data_cleaned1.rename({'Solar.R':'Solar'},axis=1)
data_cleaned3
data_cleaned3['Ozone'].hist()
```

```python
data_cleaned3.boxplot(column='Ozone')
data_cleaned3['Ozone'].describe()
```

```python
data_cleaned3['Month'].value_counts().plot.bar()
```

```python
import seaborn as sns
cols=data_cleaned3.columns
colours=['#000099','#ffff00']
sns.heatmap(data_cleaned3[cols].isnull(),cmap=sns.color_palette(colours))
```

```python
data_cleaned3[data_cleaned3.isnull().any(axis=1)].head()
```

```python
data_cleaned3.isnull().sum()
```

```python
mean=data_cleaned3['Ozone'].mean()
print(mean)
```

```python
data_cleaned3['Ozone']=data_cleaned3['Ozone'].fillna(mean)
```

```python
obj_columns=data_cleaned3[['Weather']]
```

```python
obj_columns=obj_columns.fillna(obj_columns.mode().iloc[0])
```

```python
obj_columns=obj_columns.isnull().sum()
```

```python
obj_columns
```

```python
data_cleaned3['Solar']=data_cleaned3['Solar'].fillna(mean)
data_cleaned3=data_cleaned3.drop('Weather',axis=1)
data_cleaned3
```

```python
data_cleaned3.isnull().sum()
```

```python
data_cleaned3
```

```python
data_cleaned3=pd.concat([data_cleaned3,obj_columns],axis=1)
data_cleaned3.isnull().sum()
```

```python
import seaborn as sns
sns.pairplot(data_cleaned3.iloc[:,0:6])
```

```python
data_cleaned3=data_cleaned3.drop('Year',axis=1)
```

```python
data_cleaned3['Month']=data_cleaned3['Month'].replace(['May'],'5')
data_cleaned3.corr()
```

```python
data_cleaned4=pd.get_dummies(data,columns=['Weather'])
data_cleaned4
```

```python
data_cleaned4=data_cleaned4.dropna()
data_cleaned4
data_cleaned4['Month']=data_cleaned4['Month'].replace(['May'],'5')
```

```python
#data_normalization

from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler
array=data_cleaned4.values
scaler=MinMaxScaler(feature_range=(0,1))
rescaledx=scaler.fit_transform(array[:,0:5])

#transformed data
set_printoptions(precision=2)
print(rescaledx[0:5,:])
```

```python
from sklearn .preprocessing import StandardScaler
array=data_cleaned4.values
scaler=StandardScaler().fit(array)
rescaledx=scaler.transform(array)
set_printoptions(precision=2)
print(rescaledx[0:5,:])
```
