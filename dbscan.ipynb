```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
print(df.head())
df
```

```python
print(df.info())
```

```python
df1=df.drop(['variety'],axis=1)
```

```python
array=df1.values
array
```

```python
stscaler=StandardScaler().fit(array)
x=stscaler.transform(array)
x
```

```python
dbscan=DBSCAN(eps=0.475,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
pd.set_option('display.max_rows',None)
```

```python
df1=pd.concat([df,cl],axis=1)
df1
```

```python
plt.style.use('classic')
```

```python
df1.plot(x='petal.length',y='petal.width',c=dbscan.labels_,kind='scatter',s=50,cmap=plt.cm.spring)
plt.title('clusters using DBSCAN')
```

```python
dl=dbscan.labels_
```

```python
import sklearn
sklearn.metrics.silhouette_score(x,dl)
```

```python
from sklearn.cluster import KMeans
clf=KMeans(n_clusters=3)
y_kmeans=clf.fit_predict(x)
```

```python
y_kmeans
```

```python
cl1=pd.DataFrame(y_kmeans,columns=['kcluster'])
cl1
```

```python
df2=pd.concat([df1,cl1],axis=1)
df2
```

```python
df2.plot(x='petal.length',y='petal.width',c=y_kmeans,kind='scatter',s=50,cmap=plt.cm.spring)
plt.title('cluster using kmeans')
```

```python
sklearn.metrics.silhouette_score(x,y_kmeans)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh=NearestNeighbors(n_neighbors=8)
nbrs=neigh.fit(x)
distances,indices=nbrs.kneighbors(x)
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```
