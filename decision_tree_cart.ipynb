```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
iris=pd.read_csv("iris.csv")
iris
```

```python
#labelling each variety with numbers
label_encoder=preprocessing.LabelEncoder()
iris['variety']=label_encoder.fit_transform(iris['variety'])
```

```python
x=iris.iloc[:,0:4]
y=iris['variety']
```

```python
x
pd.set_option("display.max_rows",None)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
#plot decision tree
tree.plot_tree(model);
```

```python
fn=['speal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)']
cn=['setosa','versicolor','virginia']
#ncols and nrows by default it takes as 1
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True);
```

```python
text=tree.export_text(model,max_depth=3)
print(text)
```

```python
model.feature_importances_
```

```python
import pandas as pd
feature_imp=pd.Series(model.feature_importances_,index=fn).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x=feature_imp,y=feature_imp.index)
plt.xlabel('Feature importance score')
plt.ylabel('Features')
plt.title('Visualizing important features')
plt.show()
```

```python
preds=model.predict(x_test)
pd.Series(preds).value_counts()
```

```python
pd.crosstab(y_test,preds)
```

```python
#finding accuracy
np.mean(preds==y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
array=iris.values
x=array[:,0:3]
y=array[:,3]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
model=DecisionTreeRegressor(max_depth=3)
model.fit(x_train,y_train)
```

```python
#finding testing accuracy
model.score(x_test,y_test)
```

```python
#finding training accuracy
model.score(x_train,y_train)
```
