```python
import pandas as pd
import numpy as np
```

```python
beml_df=pd.read_csv('BEML.csv')
```

```python
beml_df[0:5]
```

```python
glaso_df=pd.read_csv('GLAXO.csv')
```

```python
glaso_df[0:5]
```

```python
beml_df=beml_df[['Date','Close']]
```

```python
glaso_df=glaso_df[['Date','Close']]
```

```python
beml_df
```

```python
glaso_df
```

```python
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
beml_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(glaso_df.Close)
plt.xlabel('Time')
plt.ylabel('Close price')
plt.show()
```

```python
glaso_df['gain']=glaso_df.Close.pct_change(periods=1)
beml_df['gain']=beml_df.Close.pct_change(periods=1)
```

```python
beml_df
```

```python
glaso_df
```

```python
glaso_df=glaso_df.dropna()
beml_df=beml_df.dropna()
```

```python
glaso_df.isnull().sum()
```

```python
beml_df.isnull().sum()
```

```python
plt.figure(figsize=(8,6))
plt.plot(glaso_df.gain)
plt.show()
```

```python
sn.distplot(glaso_df.gain,label='glaso')
plt.show()
```

```python
sn.distplot(glaso_df.gain,label='glaso')
plt.legend()
plt.show()
```

```python
sn.distplot(beml_df.gain,label='beml',color='r')
plt.legend()
plt.show()
```

```python
x=pd.DataFrame(columns=["beml_df.gain","glaso_df.gain"])
x['beml_df.gain']=pd.Series(beml_df.gain)
x['glaso_df.gain']=pd.Series(glaso_df.gain)
```

```python
sn.distplot(x['glaso_df.gain'],label='Glaso')
sn.distplot(x['beml_df.gain'],label='beml')
plt.legend()
plt.show()
```

```python
glaso_df['gain'].mean()
```

```python
beml_df['gain'].mean()
```

```python
print('mean:',np.round(glaso_df.gain.mean(),4))
print('standard deviation:',np.round(glaso_df['gain'].std(),4))
```

```python
from scipy import stats
stats.norm.cdf(-0.02,glaso_df.gain.mean(),glaso_df.gain.std())
#this is for probabiity of making 2% loss or higher in glaso
```

```python
1-stats.norm.cdf(0.02,glaso_df.gain.mean(),glaso_df.gain.std())
#this is for probability of making 2% gain or higher in glaso
```
